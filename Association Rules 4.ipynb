```python
import numpy as np
import pandas as pd

from mlxtend.frequent_patterns import apriori
from mlxtend.frequent_patterns import association_rules
```

```python
association=pd.read_csv('association.csv', low_memory=False)
```

```python
association=association.drop_duplicates()
```

```python
association.shape
```

```python
finalData = association.groupby(['UNIQUETRANS']).sum()
```

```python
finalData.head(5)
```

```python
frequentItemsets = apriori(finalData, min_support=0.001, use_colnames=True)
```

```python
rules = association_rules(frequentItemsets, metric="lift", min_threshold=1)
```

```python
print(rules)
```

```python
rules.describe()
```

```python
rules[ rules['lift'] >= 990].describe()
```

```python
rules=rules[(rules['lift'] >= 900) & (rules['confidence'] >= 1)]
```

```python
rules.describe()
```

```python
export_csv = rules.to_csv ('rules.csv', index = None, header=True)
```
